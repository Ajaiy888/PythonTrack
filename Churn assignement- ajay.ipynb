```python
#c.	Extract all the male senior citizens whose Payment Method is Electronic check & store the result in â€˜senior_male_electronicâ€™
senior_male_electronic = df[(df['gender'] == 'Male') & 
                            (df['SeniorCitizen'] == 1) & 
                            (df['PaymentMethod'] == 'Electronic check')]
print(senior_male_electronic)
```

```python
#d.Extract all those customers whose tenure is greater than 70 months or their Monthly charges is more than 100$ & store the result in â€˜customer_total_tenureâ€™
customer_total_tenure = df[(df['tenure']>70.0 )& (df['MonthlyCharges']>100.0)]
print(customer_total_tenure)
```

```python
#e.	Extract all the customers whose Contract is of two years, payment method is Mailed check & the value of Churn is â€˜Yesâ€™ & store the result in â€˜two_mail_yesâ€™
two_mail_yes = df[(df['Contract']=='Two year')&(df['PaymentMethod']=='Mailed check')& (df['Churn']=='Yes')]
print(two_mail_yes)
```

```python
#f.	Extract 333 random records from the customer_churn dataframe & store the result in â€˜customer_333â€™



customer_333 = df.sample(n=333, random_state=42)  


print(customer_333.head(5))
```

```python
# g.Get the count of different levels in the 'Churn' column
churn_counts = df['Churn'].value_counts()

print(churn_counts)
```

```python
'''a.	Build a bar-plot for the â€™InternetServiceâ€™ column:
i.	Set x-axis label to â€˜Categories of Internet Serviceâ€™
ii.	Set y-axis label to â€˜Count of Categoriesâ€™
iii.	Set the title of plot to be â€˜Distribution of Internet Serviceâ€™
iv.	Set the color of the bars to be â€˜orangeâ€™
'''
import pandas as pd
import matplotlib.pyplot as plt

internet_service_counts = df['InternetService'].value_counts()
print(internet_service_counts)

plt.bar(internet_service_counts.index, internet_service_counts.values, color='orange')
plt.xlabel('Categories of Internet Service')

plt.ylabel('Count of Categories')

plt.title('Distribution of Internet Service')

plt.show()
```

```python
'''b.	Build a histogram for the â€˜tenureâ€™ column:
i.	Set the number of bins to be 30
ii.	Set the color of the bins  to be â€˜greenâ€™
iii.	Assign the title â€˜Distribution of tenureâ€™'''
import pandas as pd
import matplotlib.pyplot as plt

tenure_data = df['tenure']  

plt.hist(tenure_data, bins=30, color='green', edgecolor='black')

plt.title('Distribution of Tenure')

plt.show()

```

```python
'''c.	Build a scatter-plot between â€˜MonthlyChargesâ€™ & â€˜tenureâ€™. Map â€˜MonthlyChargesâ€™ to the y-axis & â€˜tenureâ€™ to the â€˜x-axisâ€™:
i.	Assign the points a color of â€˜brownâ€™
ii.	Set the x-axis label to â€˜Tenure of customerâ€™
iii.	Set the y-axis label to â€˜Monthly Charges of customerâ€™
iv.	Set the title to â€˜Tenure vs Monthly Chargesâ€™
'''
plt.scatter(df['tenure'], df['MonthlyCharges'], color='brown', s=5)

plt.xlabel('Tenure of Customer')
plt.ylabel('Monthly Charges of Customer')
plt.title('Tenure vs Monthly Charges')
plt.show()
```

```python
#d.d.	Build a box-plot between â€˜tenureâ€™ & â€˜Contractâ€™. Map â€˜tenureâ€™ on the y-axis & â€˜Contractâ€™ on the x-axis. 
df.boxplot(column='tenure', by='Contract', grid=False)
plt.suptitle('')  
plt.title('Tenure by Contract Type')
plt.xlabel('Contract')
plt.ylabel('Tenure of Customer')
```
