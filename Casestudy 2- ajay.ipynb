```python
import pandas as pd
import numpy as np

schools=pd.read_csv('middle_tn_schools.csv')
grouped_data = schools.groupby('school_rating')['reduced_lunch']


summary = grouped_data.describe()
print(summary)
schools_corr = schools[['school_rating', 'size', 'reduced_lunch', 'state_percentile_16',
                   'state_percentile_15', 'stu_teach_ratio', 'avg_score_15', 
                   'avg_score_16', 'full_time_teachers', 'percent_black', 
                   'percent_white', 'percent_asian', 'percent_hispanic']].corr()


schools_corr = schools_corr.round(2)


print(schools_corr)
```

```python

correlation = schools['reduced_lunch'].corr(schools['school_rating'])


print(f"Correlation between reduced_lunch and school_rating: {correlation:.2f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


sns.set(style="whitegrid")


plt.figure(figsize=(6, 6))
sns.regplot(x='reduced_lunch', y='school_rating', data=schools, scatter_kws={'s': 50}, line_kws={"color": "red"})


plt.xlabel('Percentage of Students on Reduced Lunch (%)')
plt.ylabel('School Rating')
plt.title('Relationship Between School Rating and Reduced Lunch')

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


corr_matrix = schools[['school_rating', 'size', 'reduced_lunch', 'state_percentile_16', 
                  'state_percentile_15', 'stu_teach_ratio', 'avg_score_15', 
                  'avg_score_16', 'full_time_teachers', 'percent_black', 
                  'percent_white', 'percent_asian', 'percent_hispanic']].corr()


plt.figure(figsize=(10, 8))


sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0, 
            annot_kws={"size": 10}, square=True, linewidths=.5, cbar_kws={"shrink": .75})

plt.title('Correlation Matrix of School Data', fontsize=15)


plt.show()
```
